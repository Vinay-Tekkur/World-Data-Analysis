```python
# Seaborn Visualization
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
# distribution plot which give more insights for InternetUsers
vis1 = sns.distplot(stats_csv['InternetUsers'], bins=100)
```

```python
# Box plot which give more insights of data between Birthrate categorized by Income level of the country
vis2 = sns.boxplot(data=stats_csv, x="IncomeGroup", y="BirthRate")
```

---

```python
# Scatter plot which give more insights of data between Birthrate by Internet Users of the country with regression line
vis3 = sns.lmplot(data=stats_csv, x="InternetUsers", y="BirthRate", fit_reg=True)
```

```python
# Scatter plot which give more insights of data between Birthrate and InternetUsers categorized by Income level of the country
vis3 = sns.lmplot(data=stats_csv, x="InternetUsers", y="BirthRate", fit_reg=False, hue='IncomeGroup', size=8)
```
